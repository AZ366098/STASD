```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
% matplotlib inline
```

```python
with open('report2', 'r') as myfile:
    demo_text=myfile.read().replace('\n', '')

stopwords = ['我', '對', '\n']
puncs = '！？｡＂＃＄％＆＇（）＊＋，－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘’‛“”„‟…‧﹏.。'
```

```python
seg_list = jieba.cut(demo_text, cut_all=False)  # 精確模式
print("Default Mode: " + "/ ".join(seg_list))  
```

```python
seg_list = jieba.cut(demo_text, cut_all=False)
seg_list = list(seg_list)
wordDict = {}
for w in seg_list:
    if (w not in stopwords) and (w not in puncs):
        if w in wordDict:
            wordDict[w] = wordDict[w] + 1
        else:
            wordDict[w] = 1
print(wordDict)
```

```python
font = r"/usr/share/fonts/truetype/wqy/wqy-microhei.ttc"
wordcloud = WordCloud(font_path=font)
wordcloud.fit_words(wordDict)
plt.imshow(wordcloud)
```

```python
plt.rcParams['font.sans-serif'] = ''
#plt.rcParams["font.sans-serif"]= 'SimHei'
```

```python
sorted_wordDict = sorted(wordDict.items(), key=lambda x: x[1], reverse=True)
max_num = 20
index = [i for i in range(max_num)]
label = [i[0] for i in sorted_wordDict[:max_num]]
value = [i[1] for i in sorted_wordDict[:max_num]]
plt.figure(figsize=(10,6))
plt.bar(index,value)
plt.xticks(index,label)
```
