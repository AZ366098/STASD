## 1. 快速認識資料

```python
import pandas as pd
```

```python
df = pd.read_csv("../91APPdataset/Orders.csv")
```

```python
df.head()
```

```python
df['new_DateId'] = pd.to_datetime(df['DateId'].astype(str), format='%Y%m%d')
```

```python
df.info()
```

```python
df.head()
```

## 2. 資料選取 (Slice) 與過濾 (Filter)

### 2.1 資料選取 (Slice)

```python
df['Quantity'].head()
```

```python
df.loc[3]
```

```python
df.loc[1000:1005, 'Quantity']
```

```python
df.loc[1000:1005, ['new_DateId', 'Quantity']]
```

### 2.2 資料過濾 (Filter)

```python
df[df['Quantity'] > 1]
```

## 3. 資料分組 (groupby)

```python
data = df.groupby(by="new_DateId")['Quantity'].sum()
data.head(10)
```

```python
%matplotlib inline
data.plot()
```
