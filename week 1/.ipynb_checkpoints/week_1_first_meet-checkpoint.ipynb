## 處理非數值資料
* 透過 Hot Key Encoding 轉換資料，加入原資料集
* 敘述統計

### TrackSourceTypeDef 訂單來源
* 從瀏覽器下單的客戶最多，達 60.8%
* 使用 Android 系統的用戶(20.3%)稍多於 iOS 的用戶(18.9%)，但差異不大

```python
source_type = pd.get_dummies(data['TrackSourceTypeDef'], prefix='Source')
data = pd.concat([data, source_type], axis=1)
```

```python
data.groupby(by='SalesOrderReceiverId')['Source_AndroidApp','Source_Web', 'Source_iOSApp'].mean().describe()
```

### TrackDeviceTypeDef 訂單裝置來源
* 透過智慧型手機購物的用戶居多，有 84.9%
* 使用電腦購物的用戶有 15.1%
* 幾乎沒有人用平板購物

```python
device_type = pd.get_dummies(data["TrackDeviceTypeDef"], prefix="Device")
data = pd.concat([data, device_type], axis=1)
```

```python
data.groupby(by='SalesOrderReceiverId')['Device_Mobile', 'Device_PC', 'Device_Pad'].mean().describe()
```

### PayProfileTypeDef 付款方式
* 90% 用戶習慣使用超商付款(7-11: 55.2% > 全家: 37.6%)

* 用 ATM 和 LinePay 的客戶非常少(0.1%)

```python
pay = pd.get_dummies(data['PayProfileTypeDef'], prefix='Pay')
data = pd.concat([data, pay], axis=1)
```

```python
data.groupby(by='SalesOrderReceiverId')['Pay_ATM', 'Pay_CreditCardOnce', 'Pay_Family', 'Pay_LinePay', 'Pay_SevenEleven'].mean().describe()
```

## 時間變數
##### 把字串型別轉換成Timestamp類別的變數
* 做出年、月、日、以及季變數

```python
data['new_DateId'] = pd.to_datetime(data['DateId'].astype(str), format='%Y%m%d')
date_quantity = data.groupby('new_DateId')['Quantity'].sum()
date_quantity.plot()
```

```python
data['Year'] = data['new_DateId'].apply(lambda x: x.year)
data['Month'] = data['new_DateId'].apply(lambda x: x.month)
data['Day'] = data['new_DateId'].apply(lambda x: x.day)
```

```python
salesorder_month = data.groupby('SalesOrderReceiverId')['Month'].mean().reset_index(name='Month')
sales_month_rank = salesorder_month.groupby('Month').size().reset_index(name='count')
sales_month_rank.sort_values('count', ascending=False)
```

```python
data['Quarter'] = data['Month'].apply(lambda x: ((x-1) - (x-1) % 3) // 3 + 1)
salesorder_quarter = data.groupby('SalesOrderReceiverId')['Quarter'].mean().reset_index(name='Quarter')
sales_quarter_rank = salesorder_quarter.groupby('Quarter').size().reset_index(name='count')
sales_quarter_rank.sort_values('count', ascending=False)
```
